```
N = [None] * 1000
T = int(input())

j = 0
for i in range(1000):
  N[i] = j
  print('N[{}]: {}'.format(i, j))
  if j < (T - 1):
    j = j + 1
  else:
    j == T - 1
    j = 0
```
