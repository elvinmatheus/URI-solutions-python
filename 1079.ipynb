```
N = int(input())
while N > 0:
  x = input().split()
  a = float(x[0])
  b = float(x[1])
  c = float(x[2])
  M = (a * 2 + b * 3 + c * 5) / 10
  N = N - 1
  print('{:.1f}'.format(M))
```
